```python
a = 17
m = 100
z0 = 42
n = 15
```

```python
results = []
```

```python
z = z0
for i in range(1, n + 1):
    z_prev = z
    z = (a * z) % m
    u = z / m  # Bilangan acak seragam
    rumus = f"(17*{z_prev})mod{m}"  # Rumus iterasi
    results.append([i, z_prev, rumus, z, round(u, 3)])
```

```python
# Header tabel
print(f"{'i':<3} | {'zi-1 (Bilangan sebelumnya)':<28} | {'Rumus zi = (17.zi-1)mod 100':<32} | {'zi (Nomor Mahasiswa)':<20} | {'ui = zi/100 (Bilangan Acak Seragam)':<20}")
print("_" * 120)

# Isi tabel
for result in results:
    print(f"{result[0]:<3} | {result[1]:<28} | {result[2]:<32} | {result[3]:<20} | {result[4]:<20}")
```

```python
import matplotlib.pyplot as plt

# Data untuk grafik
iterasi = [result[0] for result in results]
nilai_u = [result[4] for result in results]

# Plot grafik
plt.figure(figsize=(10, 5))
plt.plot(iterasi, nilai_u, marker='o', linestyle='-', color='blue', label='u_i (Bilangan Acak Seragam)')
plt.title('Grafik Bilangan Acak Seragam (u_i) Selama Iterasi')
plt.xlabel('Iterasi (i)')
plt.ylabel('u_i')
plt.grid(True)
plt.legend()
plt.show()
```
